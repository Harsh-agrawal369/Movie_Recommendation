# This a movie Recommender System. 
**It suggests you the best movie which are similar to what you like.
Here we have used the content Based approach to find movies.
We will be using vectorization to recommend movies.**

```python
#Importing Libraries

import numpy as np
import pandas as pd
```

```python
#importing Dataset

credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
#Merging of both dataFrames

movies = movies.merge(credits, on = 'title')

movies.head(2)
```

**We are not keeping duration, Popularity and Year of making beacuse the are numerical column and they won't fit in.**

```python
#Popularity and duration are not kept because they are numeric

movies = movies[['genres', 'id', 'keywords', 'overview', 'title', 'cast', 'crew']]

movies.head(2)
```

# Data Preprocessing

```python
movies.isnull().any()
```

```python
#We have null values only in Overview column.
#Hence checking the number of null values.

movies.isnull().sum()
```

**We only have 3 null values in the overview column. Hence we will remove those rows as it won't affect much.**

```python
movies.dropna(inplace=True)
```

```python
#Again checking after droping all null value entries

movies.isnull().sum()
```

```python
#Checking for duplicated data

movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

**Putting genres in proper format by merging them to a string**

```python
#Calling function Literal_eval for converting string to list

import ast
```

```python
def convert(obj):
    Lst=[]
    for i in ast.literal_eval(obj):
        Lst.append(i['name'])
    return Lst
```

**We cannot Pass the genres directly to the function beacuse genres is in the form of String. Our fucntion cannot take string object, so we have to convert string to list.**

```python
#Calling convert function for all the genres entries in the dataset

movies['genres']=movies['genres'].apply(convert)
```

**We also need to get names for the keyword columns. Hence, applying convert function to keywords as well.**

```python
#Calling convert function for all the keywords entries in the dataset

movies['keywords']=movies['keywords'].apply(convert)
```

**We will need to get best Actors name from the cast column. We will take first 4 actor's name.**

```python
def convert_cast(obj):
    counter=0
    Lst=[]
    for i in ast.literal_eval(obj):
        if counter!=4:
            Lst.append(i['name'])
            counter+=1
        else:
            break
    return Lst
```

```python
#Calling convert_cast function to the cast column

movies['cast']=movies['cast'].apply(convert_cast)
```

**We also need to get the name of director from the crew column**

```python
def get_director(obj):
    Lst=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            Lst.append(i['name'])
            break
        else:
            continue
    return Lst
```

```python
#Calling get_director function to the crew column

movies['crew']=movies['crew'].apply(get_director)
```

**Overview column is a String, converting it to list.**

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#Checking movies data columns

movies.head()
```

**We will need to remove space between words, because there may be some confusion if two enteties of same first name and different second name exist.**

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

# Creating new column tags
**This tags column containes data of columns genres, keywords, overview, cast and crew.**

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

**Dropping all the other columns**

```python
df_mov = movies[['id', 'title', 'tags']]
```

```python
df_mov.head()
```

```python
df_mov['tags'] = df_mov['tags'].apply(lambda x:" ".join(x))
```

```python
df_mov.head(2)
```

**Converting everything to LowerCase**

```python
df_mov['tags'] = df_mov['tags'].apply(lambda x:x.lower())
```

```python
df_mov.head(2)
```

**Here We have to apply stemming to the tags column, so that we do not get same words in different tenses/forms**
**We will use nltk for the same.**

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df_mov['tags'] = df_mov['tags'].apply(stem)
```

```python
df_mov.iloc[0].tags
```

# Performing Vectorization
**Bags of Words Technique is used**

```python
#Importing CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000, stop_words='english')
```

```python
#Creating Vectors

vectors = cv.fit_transform(df_mov['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors
```

```python
#Looking at all the features

cv.get_feature_names()
```

**We will use ditance to find the similarity between two vectors.**
**Since we are in multidemisinal space having 6000 dimensions, we cannot use eucledian diatance as it it not reliable in the multidimensional space, instead, we will use cosine distance.**

```python
#Importing Libraries

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity[:2]
```

**Now we need to find the movies which are most similar to the given movie. We also need to sort that list on the basis of cosine similarity and then we will fetch top 5 movies on that basis.**

```python
def recommended(movie):
    movies_index = df_mov[df_mov['title'] == movie].index[0]
    distances = similarity[movies_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df_mov.iloc[i[0]].title)
```

```python
recommended('Falcon Rising')
```
